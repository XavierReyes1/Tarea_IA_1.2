```python
import pandas as pd
```

```python
df = pd.read_csv('clima.csv')
df.head()
```

. Calcule la temperatura promedio de cada ciudad.


. Determine el registro con la temperatura más alta y el más bajo en el dataset.


. Identifique qué ciudad tuvo la temperatura más alta y cuál la más baja en todo el dataset.


. Encuentre cuántos registros tienen una temperatura mayor a 30°C.


. Cuenta cuántos días en total hay registrados por cada ciudad.


```python
#Calcule la temperatura promedio de cada ciudad.
temperatura_promedio = df.groupby('Ciudad')['Temperatura'].mean()
temperatura_promedio
```

```python
#Determine el registro con la temperatura más alta y el más bajo en el dataset.
Temperatura_alta = df[df['Temperatura'] == df['Temperatura'].max()]
print("Temperatua mas alta: ")
print(Temperatura_alta)
Temperatura_baja = df[df['Temperatura'] == df['Temperatura'].min()]
print("Temperatua mas baja: ")
print(Temperatura_baja)
```

```python
#Identifique qué ciudad tuvo la temperatura más alta y cuál la más baja en todo el dataset.

Temperatura_max_min = df[(df['Temperatura'] == df['Temperatura'].max())+(df['Temperatura'] == df['Temperatura'].min())]
Ciudad_temperatura = Temperatura_max_min.Ciudad.head(1)
print("Ciudad de mas alta y mas baja temperatura: ")
print(Ciudad_temperatura)
```

```python
#Encuentre cuántos registros tienen una temperatura mayor a 30°C.
temperatura_mayor_30 = (df['Temperatura'] > 30).sum()

print("Numero de registros mayores a 30°C son: "+ str(temperatura_mayor_30) + " Registros")
```

```python
# Cuenta cuántos días en total hay registrados por cada ciudad.

dias_registrados = df.groupby('Ciudad')['Fecha'].count()
dias_registrados = dias_registrados.rename('Dias')
dias_registrados
```
